```python
# Import the numeric library for data handling
# and the matplotlib graphics library
import matplotlib.pyplot as plot
import numpy as np
import pandas as panda

from matplotlib.ticker import AutoMinorLocator, MultipleLocator
from matplotlib.colors import LinearSegmentedColormap

from numpy.polynomial import Polynomial

# Import the Excel file handling code
from pandas import ExcelFile

# Import the statistics handling and plotting code
import statsmodels.graphics.gofplots as sm
import statsmodels.sandbox.distributions.extras as sm_extra

from scipy import stats
from scipy.stats import norm, skewnorm
```

```python
# Calculate the interval between 0 and 20 in steps of 0.001
x_axis = np.arange(0, 100, 0.01)
y_axis = np.array([150])

# Plot the result
fig, ax = plot.subplots()

# ax.plot(x_axis,y_axis)

ax.grid(True, which="both")

ax.set_title(
    "Risk Map Showing the Relationship Between Turnover and Temperature\n on Legonella Growth for a Model Water System",
    pad=24.0,
)
ax.set_xlabel("Temperature (C)")
ax.set_ylabel("Turnover (%)")

ax.set_ylim(0, 150)
ax.set_xlim(0, 90)

# Fill in the Temperature Regions of Interest
# From [Legionnaires Disease & Home Hot Water: Understanding Risks and Prevention](https://renewableheatinghub.co.uk/legionnaires-disease-home-hot-water-understanding-risks-and-prevention)
ax.fill_between(
    x_axis,
    y_axis,
    0,
    where=(x_axis < 45),
    color="orange",
)
ax.fill_between(x_axis, y_axis, 0, where=(x_axis < 42), color="red", label="High Risk")
ax.fill_between(
    x_axis, y_axis, 0, where=(x_axis < 32), color="orange", label="Medium High Risk"
)

# Fill in the Flow Regions of Interest
# From [Legionnaires Disease & Home Hot Water: Understanding Risks and Prevention](https://renewableheatinghub.co.uk/legionnaires-disease-home-hot-water-understanding-risks-and-prevention)
ax.fill_between(x_axis, 0, 25, color="red", alpha=0.5)
ax.fill_between(x_axis, 25, 100, color="orange", alpha=0.5, label="Medium Low Risk")
ax.fill_between(x_axis, 75, 150, color="green", alpha=0.8)

# Blank regions not of interest, due to low temperature
# From [Legionnaires Disease & Home Hot Water: Understanding Risks and Prevention](https://renewableheatinghub.co.uk/legionnaires-disease-home-hot-water-understanding-risks-and-prevention)
ax.fill_between(x_axis, y_axis, 0, where=(x_axis > 45), color="green", label="Low Risk")
ax.fill_between(
    x_axis,
    y_axis,
    0,
    where=(x_axis > 50),
    color="blue",
    alpha=0.6,
    label="Pasturisation",
)
ax.fill_between(x_axis, y_axis, 0, where=(x_axis > 70), color="black", label="Dead")
ax.fill_between(x_axis, y_axis, 0, where=(x_axis < 20), color="green")

# Add Labels
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, frameon=False)
```
